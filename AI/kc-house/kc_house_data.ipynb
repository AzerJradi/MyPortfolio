```python
fig = px.histogram(df,x="bedrooms" , y="price" ,  color_discrete_sequence=['red'], nbins=10, height=400)
fig.update_layout(
    title_text="Distribution of price by bedrooms",
    xaxis_title="bedrooms",
    yaxis_title="price"
)
fig.show()
```

```python
df_waterfront_sales = df.groupby('waterfront')['price'].sum().reset_index()
fig = px.pie(df_waterfront_sales, values='price', names='waterfront', height=400)
fig.update_layout(
    title_text="Distribution of waterfront by Global Sales",
    showlegend=True
)

fig.show()
```

```python
fig = px.scatter(df, x="yr_built", y="price",color_discrete_sequence=['red'], height=500)
fig.update_layout(
    title_text="Relationship Between yr_built Sales and price",
    xaxis_title="yr_built Sales",
    yaxis_title="price"
)
fig.show()
```
