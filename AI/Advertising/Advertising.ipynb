```python
import plotly.express as px
```

```python
fig = px.bar(df, x="Sales", y="TV", height=400)
fig.update_layout(
    title_text="Sales by TV Advertising Spend",
    xaxis_title="Sales",
    yaxis_title="TV Advertising Spend"
)
fig.update_traces(marker=dict(color='red', opacity=0.7, line=dict(color='red', width=1.5)))
fig.show()
```
