```python
import pandas as pd

file_path = '/content/drive/MyDrive/Video games.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df["Year"] = df["Year"].fillna(df["Year"].mean())
```

```python
df["Publisher"].value_counts()
```

```python
df["Publisher"] = df["Publisher"].fillna("Electronic Arts")
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
import plotly.express as px
```

```python
fig = px.area(df, x="Year", y="Global_Sales", height=400)
fig.update_layout(
    title_text="Global Sales by Year",
    xaxis_title="Year",
    yaxis_title="Global Sales"
)
fig.update_traces(fillcolor='red', line=dict(color='red', width=1.5), opacity=0.3)
fig.show()
```

```python
fig = px.histogram(df, y="Global_Sales",x="Year" , color_discrete_sequence=['red'], nbins=10, height=400)
fig.update_layout(
    title_text="Distribution of Global Sales by Year",
    xaxis_title="Years",
    yaxis_title="Global Sales"
)
fig.show()
```

```python
fig = px.scatter(df, x="NA_Sales", y="EU_Sales", color_discrete_sequence=['red'] , height=400)
fig.update_layout(
    title_text="Relationship Between NA_Sales Sales and EU_Sales",
    xaxis_title="NA_Sales Sales",
    yaxis_title="EU_Sales"
)
fig.show()
```

```python
fig = px.scatter(df, x="JP_Sales", y="Other_Sales",color_discrete_sequence=['red'], height=400)
fig.update_layout(
    title_text="Relationship Between JP_Sales Sales and Other_Sales",
    xaxis_title="JP_Sales Sales",
    yaxis_title="Other_Sales"
)
fig.show()
```

```python
df_pivot = df.pivot_table(index='Year', columns='Platform', values='Global_Sales', aggfunc='sum').reset_index()
fig = px.area(df_pivot, x='Year', y=df_pivot.columns[1:], height=400)
fig.update_layout(
    title_text="Distribution of Global Sales by Platform Over the Years",
    xaxis_title="Year",
    yaxis_title="Global Sales",
    legend_title="Platform"
)

fig.show()
```

```python
df_top10 = df.sort_values(by='Global_Sales', ascending=True).head(10)

# Create a bar chart
fig = px.bar(df_top10, x='Publisher', y='Global_Sales', height=400)

# Update layout for better aesthetics
fig.update_layout(
    title_text="Top 10 Publishers by Global Sales",
    xaxis_title="Publisher",
    yaxis_title="Global Sales",
    width=1450  # Adjust width as needed
)

fig.show()
```

```python
df_genre_sales = df.groupby('Genre')['Global_Sales'].sum().reset_index()
fig = px.pie(df_genre_sales, values='Global_Sales', names='Genre', height=400)
fig.update_layout(
    title_text="Distribution of Genres by Global Sales",
    showlegend=True
)

fig.show()
```
